```python
from pyTVDN import TVDNDetect
from pathlib import Path
from scipy.io import loadmat
from scipy.signal import detrend
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
resDir = Path("../results")
dataDir = Path("../data")
```

## 1 Data 02

```python
filname = Path("./MEG-eyes10092020/lcmv_result_R2868E_02-st.mat")
MEG = loadmat(dataDir/filname)
```

```python
data = MEG["sa"][0][0]
```

```python
data.shape
```

The time course is 60 seconds, so the frequency is 1200 Hz

```python
d, n = data.shape
```

### Four sample sequences of the data

```python
plt.figure(figsize=(10, 10))
plt.subplot(221)
plt.plot(data[1, :])

plt.subplot(222)
plt.plot(data[25, :])

plt.subplot(223)
plt.plot(data[50, :])

plt.subplot(224)
plt.plot(data[100, :])
```

### The first 100 seqs

```python
for i in range(100):
    plt.plot(data[i, :])
```

###  Data preprocessing

```python
from scipy.signal import lfilter, firls, decimate
```

```python
dataN = decimate(data, q=10)
```

```python
dataN = detrend(dataN)
```

### HPF 

```python
fs = 120 #sampling frequency
fmin = 2 # 2Hz - 45Hz signal range, filter for this with hbp
fmax = 45
hbp = firls(101, np.array([0, 0.2*fmin, 0.9*fmin, fmax-2, fmax+5, 60])*2/fs,
           desired = np.array([0, 0, 1, 1, 0, 0])) #for detrending, a bandpass
```

```python
dataHPF = lfilter(hbp, 1, dataN)
```

### Four sequences of the data after HPF

```python
plt.figure(figsize=(10, 10))
plt.subplot(221)
plt.plot(dataHPF[1, :])

plt.subplot(222)
plt.plot(dataHPF[25, :])

plt.subplot(223)
plt.plot(dataHPF[50, :])

plt.subplot(224)
plt.plot(dataHPF[100, :])
```

### The first 100 sequences after HPF

```python
for i in range(100):
    plt.plot(dataHPF[i, :])
```
